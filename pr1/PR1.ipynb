### Вывод графиков целевых полей

```python
# Поле для прогнозирования должно быть выведено в виде линейного графика
# Не забудьте подписать!
# Замените totlngth на ваше поле для прогнозирования
plt.title('Длина хвоста', fontdict={'fontsize': 20})
plt.ylabel('Длина. см.')
plt.xlabel('Номер записи')
plt.plot(df.totlngth)
plt.show()
```

```python
#Поле для классификации надо отобразить в виде гистограммы
#Чтобы узнать число и значения классов используйте:
print(df['age'].unique())
#Замените age на ваше поле для классификации
plt.title('Распределение возраста', fontdict={'fontsize': 20})
plt.xlabel('Возраст')
plt.ylabel('Число')
#Теперь рисуем распределение - гистограмму, со значением количества опосуммов каждого возраста, а по оси x - все возраста
# Если у вас поле для классификации содержит цифры, надо их отсортировать и убрать ошибочные значения
# Вместо case подставьте любое поле, вместо age - поле для классификации
plt.bar(np.sort(df[(df['age']>0)].age.unique()),df.groupby(df.age).case.count())
plt.xticks(np.sort(df[(df['age']>0)].age.unique()))
#Если же там текст, то уберите верхние две строчки и раскомментируйте (уберите #) две нижние:
#plt.bar(df.age.unique()),df.groupby(df.age).case.count())
#plt.xticks(df.age.unique()))
#Отобразим график
plt.show()
```
